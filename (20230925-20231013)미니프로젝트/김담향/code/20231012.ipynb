```python
# pandas패키지 사용하기
import pandas as pd

# matplotlib패키지 사용하기
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Gulim'   #한글글씨를 인지할 수 있도록 글씨체 지정

# numpy패키지 사용하기
import numpy as np
```

```python
# csv파일 불러오기
df_ce = pd.read_csv('월별(2017~2022).csv', encoding = 'euc-kr')
```

```python
df_ce
```

```python
df_ce.columns
```

```python
df_ce_temp1 = df_ce.loc[3:, ['월별(1)', '2017.3', '2018.3', '2019.3', '2020.3']].astype({'2017.3': int, '2018.3':int, '2019.3' :int, '2020.3':int})
df_ce_temp1
```

```python
df_ce_temp1['2017-2018.3'] = df_ce_temp1.loc[:, ['2017.3', '2018.3']].mean(axis =1)
```

```python
df_ce_temp1
```

```python
df_ce_temp1.dtypes
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(df_ce_temp1['월별(1)'], df_ce_temp1['2019.3'], label = '2019년')
ax.plot(df_ce_temp1['월별(1)'], df_ce_temp1['2020.3'], label = '2020년')
plt.xlabel('월')
plt.ylabel('서비스업 및 기타 전력판매량(MWh)')
fig.legend()
plt.grid(axis='x')
fig.show()
```

```python
df_p= pd.read_csv('population.csv')
df_c1= pd.read_csv('consumption_1person.csv')
df_l = pd.read_csv('local.csv')
```

```python
df_p
```

```python
df_p_tmp1=pd.merge(df_p,df_l,how='left',on='local_id')
df_c1_tmp1=pd.merge(df_c1,df_l,how='left',on='local_id')
```

```python
df_c1_tmp1
```

```python
df_p_tmp1_2015 = df_p_tmp1.loc[df_p_tmp1['연도'] == 2015].sort_values(by=["location"], ascending=[True])
df_p_tmp1_2016 = df_p_tmp1.loc[df_p_tmp1['연도'] == 2016].sort_values(by=["location"], ascending=[True])
df_p_tmp1_2017 = df_p_tmp1.loc[df_p_tmp1['연도'] == 2017].sort_values(by=["location"], ascending=[True])
df_p_tmp1_2018 = df_p_tmp1.loc[df_p_tmp1['연도'] == 2018].sort_values(by=["location"], ascending=[True])
df_p_tmp1_2019 = df_p_tmp1.loc[df_p_tmp1['연도'] == 2019].sort_values(by=["location"], ascending=[True])
df_p_tmp1_2020 = df_p_tmp1.loc[df_p_tmp1['연도'] == 2020].sort_values(by=["location"], ascending=[True])
df_p_tmp1_2021 = df_p_tmp1.loc[df_p_tmp1['연도'] == 2021].sort_values(by=["location"], ascending=[True])
df_p_tmp1_2022 = df_p_tmp1.loc[df_p_tmp1['연도'] == 2022].sort_values(by=["location"], ascending=[True])
```

```python
df_c1_tmp1_2017 = df_c1_tmp1.loc[df_p_tmp1['연도'] == 2015].sort_values(by=["location"], ascending=[True])
```

```python
x = np.arange(17)
location = df_p_tmp1_2019.location
values = df_p_tmp1_2019.평균연령

plt.bar(x, values)
plt.xticks(x, location)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(df_p_tmp1_2017['location'],df_p_tmp1_2017['평균연령'], label = '2017')
ax.plot(df_p_tmp1_2018['location'],df_p_tmp1_2018['평균연령'], label = '2018')
ax.plot(df_p_tmp1_2019['location'],df_p_tmp1_2019['평균연령'], label = '2019')
ax.plot(df_p_tmp1_2020['location'],df_p_tmp1_2020['평균연령'], label = '2020')
ax.legend()
plt.xlabel('시도별')
plt.ylabel('평균연령(세)')
plt.grid(axis='x')
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(8,5))
ax.plot(df_p_tmp1_2017['location'],df_p_tmp1_2017['평균연령'], label = '2017_평균연령')
ax.plot(df_c1_tmp1_2017['location'],df_c1_tmp1_2017['인구당 주거용 전력소비량'], label = '2017_인구당 주거용 전력판매량')
ax.plot(df_c1_tmp1_2017['location'],df_c1_tmp1_2017['인구당 업무용 전력소비량'], label = '2017_인구당 업무용 전력판매량')
ax.plot(df_c1_tmp1_2017['location'],df_c1_tmp1_2017['인구당 산업용 전력소비량'], label = '2017_인구당 산업용 전력판매량')
ax.plot(df_c1_tmp1_2017['location'],df_c1_tmp1_2017['인구당 전체 전력소비량'], label = '2017_인구당 전체 전력판매량')
plt.xlabel('시도별')
plt.ylabel('전력판매량')
plt.grid(axis='x')
ax.legend()
fig.show()
```
